```python
import numpy as np
import scipy.stats as si
S, K, T, r, sigma = 90, 100, 1, 0.01, 0.3
d1 = (np.log(S/K) + (r + 0.5*sigma**2)) / (sigma*np.sqrt(T))
d2 = (np.log(S/K) + (r - 0.5*sigma**2)) / (sigma*np.sqrt(T))
Call = S*si.norm.cdf(d1,0,1) - (K*np.exp(-r*T))*si.norm.cdf(d2)
#Put = Call + (K*np.exp(-r*T)) - S  # by Putt-Call parity : S + P = C + PV(K)
print('Call Price is', round(Call,2))
#print('Put price is', round(Put,2))
```

```python
import numpy as np
import scipy.stats as si
import matplotlib.pyplot as plt

def option():
    T = 1.0
    T_list = []
    C_list = []
    while T > 0:
        T = round(T,1)
        S, K, r, sigma = 90, 100, 0.01, 0.3
        d1 = (np.log(S/K) + (r + 0.5*sigma**2)) / (sigma*np.sqrt(T))
        d2 = (np.log(S/K) + (r - 0.5*sigma**2)) / (sigma*np.sqrt(T))
        Call = S*si.norm.cdf(d1,0,1) - (K*np.exp(-r*T))*si.norm.cdf(d2)
        #Put = Call + (K*np.exp(-r*T)) - S  # by Putt-Call parity : S + P = C + PV(K)
        print('When T is',T,', the Call Price is', round(Call,2))
        #print('Put price is', round(Put,2))
        T_list.append(T)
        C_list.append(round(Call,2))
        T -= 0.1
    plt.figure(figsize=(8,5))
    plt.plot(T_list, C_list, marker='o')
    plt.title('Call Option Price vs Time to Maturity (T)')
    plt.xlabel('Time to Maturity (T)')
    plt.ylabel('Call Option Price')
    plt.grid(True)
    plt.show()
    return
option()

```
