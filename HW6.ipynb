```python
import pandas as pd
url = 'https://raw.githubusercontent.com/ChangWei614/2025fin/refs/heads/main/mediatek.csv'
df = pd.read_csv(url)
df['Close'] = df['Close'].iloc[2:].reset_index(drop=True)
for i in range(len(df["Close"])-2):
  print('從2025年起2454.tw在第',i+1,'個交易日的收盤價為',df['Close'][i])
```
